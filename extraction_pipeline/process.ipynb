```python
import os
import json
import pandas as pd

# Directory containing the ace_comment json files
comments_dir = '/Users/arnav/Desktop/nyt_pieces/extraction_pipeline/ace_comments/181'

# Collect all JSON files in the directory
json_files = sorted([f for f in os.listdir(comments_dir) if f.endswith('.json')])

rows = []
for json_file in json_files:
    with open(os.path.join(comments_dir, json_file), 'r', encoding='utf-8') as f:
        comment = json.load(f)
        
    article_id = comment.get('article_id', None)
    comment_index = comment.get('comment_index', None)
    # image_path is not always present in ace_comments, so handle with get (or set to None)
    image_path = comment.get('image_path', None)
    ace_sentences = comment.get('ace_sentences', [])

    for sent in ace_sentences:
        rows.append({
            "article_id": article_id,
            "comment_index": comment_index,
            "image_path": image_path,
            "ace_sentence": sent,
            "json_file": json_file,
        })

df = pd.DataFrame(rows)
df.to_csv("./data/ace_sentences_181.csv", index=False)
# One row per ace sentence
display(df)
```
